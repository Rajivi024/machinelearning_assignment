```python
# ============================================================
# CNN Feature Learning Tutorial Code
# Topic:
# Understanding Hierarchical Feature Learning in Convolutional
# Neural Networks Through Visualization and Experimentation
# ============================================================

# ---------------------------
# 1. Imports and basic setup
# ---------------------------

import os
import random
from dataclasses import dataclass

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.optim as optim

from torchvision import datasets, transforms
from torch.utils.data import DataLoader, random_split


# ---------------------------------------------------
# 2. Reproducibility and output folder preparation
# ---------------------------------------------------

def set_seed(seed: int = 42) -> None:
    """
    Set random seeds so results are more reproducible.
    This does not guarantee perfect reproducibility on every system,
    but it makes the training process much more stable.
    """
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

    # These two lines make cuDNN more deterministic.
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False


set_seed(42)

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {DEVICE}")

# Create folders to save outputs.
os.makedirs("figures", exist_ok=True)
os.makedirs("results", exist_ok=True)


# ---------------------------------------------------
# 3. Configuration for the experiments
# ---------------------------------------------------

@dataclass
class Config:
    batch_size: int = 128
    learning_rate: float = 1e-3
    num_epochs: int = 10
    train_val_split: float = 0.9
    num_workers: int = 2


config = Config()


# ---------------------------------------------------
# 4. Data loading
# ---------------------------------------------------

# Normalisation values commonly used for CIFAR-10.
# These help training behave more smoothly.
CIFAR10_MEAN = (0.4914, 0.4822, 0.4465)
CIFAR10_STD = (0.2470, 0.2435, 0.2616)

train_transform = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize(CIFAR10_MEAN, CIFAR10_STD),
])

test_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(CIFAR10_MEAN, CIFAR10_STD),
])

# Download CIFAR-10.
full_train_dataset = datasets.CIFAR10(
    root="./data",
    train=True,
    download=True,
    transform=train_transform
)

test_dataset = datasets.CIFAR10(
    root="./data",
    train=False,
    download=True,
    transform=test_transform
)

# Split the training set into train and validation sets.
train_size = int(len(full_train_dataset) * config.train_val_split)
val_size = len(full_train_dataset) - train_size

train_dataset, val_dataset = random_split(
    full_train_dataset,
    [train_size, val_size],
    generator=torch.Generator().manual_seed(42)
)

# Validation data should not use random augmentation.
# We replace the transform for validation with the test transform.
val_dataset.dataset.transform = test_transform

train_loader = DataLoader(
    train_dataset,
    batch_size=config.batch_size,
    shuffle=True,
    num_workers=config.num_workers
)

val_loader = DataLoader(
    val_dataset,
    batch_size=config.batch_size,
    shuffle=False,
    num_workers=config.num_workers
)

test_loader = DataLoader(
    test_dataset,
    batch_size=config.batch_size,
    shuffle=False,
    num_workers=config.num_workers
)

CLASS_NAMES = test_dataset.classes
print("Classes:", CLASS_NAMES)


# ---------------------------------------------------
# 5. Helper functions for plotting and inspection
# ---------------------------------------------------

def imshow_tensor(image_tensor, title=None, save_path=None):
    """
    Show a single image tensor after undoing normalisation.
    This is useful when we want to display a CIFAR-10 image
    in a human-readable form.
    """
    image = image_tensor.detach().cpu().clone()

    # Undo normalisation.
    mean = torch.tensor(CIFAR10_MEAN).view(3, 1, 1)
    std = torch.tensor(CIFAR10_STD).view(3, 1, 1)
    image = image * std + mean
    image = torch.clamp(image, 0, 1)

    plt.figure(figsize=(4, 4))
    plt.imshow(np.transpose(image.numpy(), (1, 2, 0)))
    plt.axis("off")
    if title:
        plt.title(title)
    if save_path:
        plt.savefig(save_path, bbox_inches="tight", dpi=300)
    plt.show()


def count_parameters(model: nn.Module) -> int:
    """
    Count the number of trainable parameters in a model.
    This is useful for comparing model complexity.
    """
    return sum(p.numel() for p in model.parameters() if p.requires_grad)


# ---------------------------------------------------
# 6. CNN model definition
# ---------------------------------------------------

class SimpleCNN(nn.Module):
    """
    A compact CNN designed for teaching and visualisation.

    The model can be configured to:
    - use either 3x3 or 5x5 filters
    - include pooling or remove pooling

    This makes it easy to run fair architecture comparisons.
    """

    def __init__(self, num_classes=10, filter_size=3, use_pooling=True):
        super().__init__()

        if filter_size not in [3, 5]:
            raise ValueError("filter_size must be 3 or 5")

        padding = filter_size // 2
        self.use_pooling = use_pooling

        self.conv1 = nn.Conv2d(3, 32, kernel_size=filter_size, padding=padding)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=filter_size, padding=padding)
        self.conv3 = nn.Conv2d(64, 128, kernel_size=filter_size, padding=padding)

        self.relu = nn.ReLU()
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)

        # Output size depends on whether pooling is used.
        # CIFAR-10 images are 32x32.
        if use_pooling:
            # Pool after conv1 and conv2:
            # 32x32 -> 16x16 -> 8x8
            flattened_dim = 128 * 8 * 8
        else:
            # No pooling:
            # stays at 32x32
            flattened_dim = 128 * 32 * 32

        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(flattened_dim, 256),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, num_classes)
        )

    def forward(self, x):
        # First convolution block
        x = self.relu(self.conv1(x))
        if self.use_pooling:
            x = self.pool(x)

        # Second convolution block
        x = self.relu(self.conv2(x))
        if self.use_pooling:
            x = self.pool(x)

        # Third convolution block
        x = self.relu(self.conv3(x))

        # Classification head
        x = self.classifier(x)
        return x


# ---------------------------------------------------
# 7. Training and evaluation functions
# ---------------------------------------------------

def train_one_epoch(model, loader, criterion, optimizer, device):
    """
    Train the model for one epoch and return average loss and accuracy.
    """
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in loader:
        images = images.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()

        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * images.size(0)

        _, predicted = torch.max(outputs, dim=1)
        correct += (predicted == labels).sum().item()
        total += labels.size(0)

    epoch_loss = running_loss / total
    epoch_acc = correct / total
    return epoch_loss, epoch_acc


@torch.no_grad()
def evaluate(model, loader, criterion, device):
    """
    Evaluate the model without updating weights.
    This is used for validation and test performance.
    """
    model.eval()
    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in loader:
        images = images.to(device)
        labels = labels.to(device)

        outputs = model(images)
        loss = criterion(outputs, labels)

        running_loss += loss.item() * images.size(0)

        _, predicted = torch.max(outputs, dim=1)
        correct += (predicted == labels).sum().item()
        total += labels.size(0)

    epoch_loss = running_loss / total
    epoch_acc = correct / total
    return epoch_loss, epoch_acc


def train_model(model, train_loader, val_loader, config, device, model_name="model"):
    """
    Full training loop for a model.
    Returns both the trained model and a history dictionary.
    """
    model = model.to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=config.learning_rate)

    history = {
        "epoch": [],
        "train_loss": [],
        "train_acc": [],
        "val_loss": [],
        "val_acc": []
    }

    for epoch in range(config.num_epochs):
        train_loss, train_acc = train_one_epoch(
            model, train_loader, criterion, optimizer, device
        )
        val_loss, val_acc = evaluate(
            model, val_loader, criterion, device
        )

        history["epoch"].append(epoch + 1)
        history["train_loss"].append(train_loss)
        history["train_acc"].append(train_acc)
        history["val_loss"].append(val_loss)
        history["val_acc"].append(val_acc)

        print(
            f"{model_name} | Epoch {epoch + 1:02d}/{config.num_epochs} | "
            f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f} | "
            f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f}"
        )

    return model, history


# ---------------------------------------------------
# 8. Plot training curves
# ---------------------------------------------------

def plot_training_curves(history, model_name, save_prefix):
    """
    Save loss and accuracy curves as separate figures.
    Separate plots are easier to read in the final report.
    """
    epochs = history["epoch"]

    # Loss curve
    plt.figure(figsize=(7, 5))
    plt.plot(epochs, history["train_loss"], label="Training loss")
    plt.plot(epochs, history["val_loss"], label="Validation loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title(f"Loss Curve: {model_name}")
    plt.legend()
    plt.tight_layout()
    plt.savefig(f"figures/{save_prefix}_loss_curve.png", dpi=300)
    plt.show()

    # Accuracy curve
    plt.figure(figsize=(7, 5))
    plt.plot(epochs, history["train_acc"], label="Training accuracy")
    plt.plot(epochs, history["val_acc"], label="Validation accuracy")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.title(f"Accuracy Curve: {model_name}")
    plt.legend()
    plt.tight_layout()
    plt.savefig(f"figures/{save_prefix}_accuracy_curve.png", dpi=300)
    plt.show()


# ---------------------------------------------------
# 9. Visualise first-layer filters
# ---------------------------------------------------

def visualize_first_layer_filters(model, save_path, max_filters=16):
    """
    Plot the learned filters from the first convolutional layer.

    For RGB filters, each kernel has 3 channels.
    We rescale values to [0, 1] for display.
    """
    weights = model.conv1.weight.detach().cpu()

    num_filters = min(max_filters, weights.shape[0])
    cols = 4
    rows = int(np.ceil(num_filters / cols))

    plt.figure(figsize=(10, 8))

    for i in range(num_filters):
        filt = weights[i]

        # Normalise each filter for display.
        filt_min = filt.min()
        filt_max = filt.max()
        filt = (filt - filt_min) / (filt_max - filt_min + 1e-8)

        filt_np = np.transpose(filt.numpy(), (1, 2, 0))

        plt.subplot(rows, cols, i + 1)
        plt.imshow(filt_np)
        plt.axis("off")
        plt.title(f"F{i+1}", fontsize=10)

    plt.suptitle("First-Layer Learned Filters", fontsize=14)
    plt.tight_layout()
    plt.savefig(save_path, dpi=300, bbox_inches="tight")
    plt.show()


# ---------------------------------------------------
# 10. Feature map extraction
# ---------------------------------------------------

def get_feature_maps(model, image_tensor, device):
    """
    Return feature maps from each convolutional stage.

    We manually step through the model so we can capture
    intermediate activations in a clear and controlled way.
    """
    model.eval()

    with torch.no_grad():
        x = image_tensor.unsqueeze(0).to(device)

        # First stage
        x1 = model.relu(model.conv1(x))
        out1 = model.pool(x1) if model.use_pooling else x1

        # Second stage
        x2 = model.relu(model.conv2(out1))
        out2 = model.pool(x2) if model.use_pooling else x2

        # Third stage
        x3 = model.relu(model.conv3(out2))

    return {
        "layer1": x1.squeeze(0).cpu(),
        "layer2": x2.squeeze(0).cpu(),
        "layer3": x3.squeeze(0).cpu()
    }


def plot_feature_maps(feature_maps, layer_name, save_path, max_maps=16):
    """
    Display a grid of feature maps for one selected layer.
    We use grayscale because it makes activations easier to compare.
    """
    fmap = feature_maps[layer_name]
    num_maps = min(max_maps, fmap.shape[0])

    cols = 4
    rows = int(np.ceil(num_maps / cols))

    plt.figure(figsize=(10, 8))
    for i in range(num_maps):
        plt.subplot(rows, cols, i + 1)
        plt.imshow(fmap[i], cmap="gray")
        plt.axis("off")
        plt.title(f"M{i+1}", fontsize=10)

    plt.suptitle(f"Feature Maps: {layer_name}", fontsize=14)
    plt.tight_layout()
    plt.savefig(save_path, dpi=300, bbox_inches="tight")
    plt.show()


# ---------------------------------------------------
# 11. Run baseline experiment
# ---------------------------------------------------

baseline_model = SimpleCNN(filter_size=3, use_pooling=True)
print("Baseline parameter count:", count_parameters(baseline_model))

baseline_model, baseline_history = train_model(
    baseline_model,
    train_loader,
    val_loader,
    config,
    DEVICE,
    model_name="Baseline CNN (3x3, pooling)"
)

plot_training_curves(
    baseline_history,
    model_name="Baseline CNN (3x3, pooling)",
    save_prefix="baseline"
)

criterion = nn.CrossEntropyLoss()
baseline_test_loss, baseline_test_acc = evaluate(
    baseline_model,
    test_loader,
    criterion,
    DEVICE
)

print(f"Baseline test loss: {baseline_test_loss:.4f}")
print(f"Baseline test accuracy: {baseline_test_acc:.4f}")


# ---------------------------------------------------
# 12. Visualise one sample image and its feature maps
# ---------------------------------------------------

# Take one sample from the test set for visualisation.
sample_image, sample_label = test_dataset[0]
sample_label_name = CLASS_NAMES[sample_label]

imshow_tensor(
    sample_image,
    title=f"Sample input image: {sample_label_name}",
    save_path="figures/sample_input_image.png"
)

baseline_feature_maps = get_feature_maps(baseline_model, sample_image, DEVICE)

plot_feature_maps(
    baseline_feature_maps,
    layer_name="layer1",
    save_path="figures/feature_maps_layer1.png"
)

plot_feature_maps(
    baseline_feature_maps,
    layer_name="layer2",
    save_path="figures/feature_maps_layer2.png"
)

plot_feature_maps(
    baseline_feature_maps,
    layer_name="layer3",
    save_path="figures/feature_maps_layer3.png"
)


# ---------------------------------------------------
# 13. Visualise first-layer filters
# ---------------------------------------------------

visualize_first_layer_filters(
    baseline_model,
    save_path="figures/first_layer_filters.png",
    max_filters=16
)


# ---------------------------------------------------
# 14. Experiment: compare filter sizes
# ---------------------------------------------------

filter_results = []

for filter_size in [3, 5]:
    print(f"\nRunning filter-size experiment with {filter_size}x{filter_size} filters")

    model = SimpleCNN(filter_size=filter_size, use_pooling=True)

    trained_model, history = train_model(
        model,
        train_loader,
        val_loader,
        config,
        DEVICE,
        model_name=f"CNN ({filter_size}x{filter_size}, pooling)"
    )

    test_loss, test_acc = evaluate(trained_model, test_loader, criterion, DEVICE)

    filter_results.append({
        "model_name": f"CNN ({filter_size}x{filter_size}, pooling)",
        "filter_size": f"{filter_size}x{filter_size}",
        "pooling": "Yes",
        "parameters": count_parameters(trained_model),
        "test_loss": test_loss,
        "test_accuracy": test_acc,
        "main_observation": ""
    })

filter_results_df = pd.DataFrame(filter_results)
print("\nFilter size comparison:")
print(filter_results_df)

filter_results_df.to_csv("results/filter_size_results.csv", index=False)

# Plot test accuracy comparison for filter sizes.
plt.figure(figsize=(7, 5))
plt.bar(filter_results_df["filter_size"], filter_results_df["test_accuracy"])
plt.xlabel("Filter size")
plt.ylabel("Test accuracy")
plt.title("Effect of Filter Size on Test Accuracy")
plt.tight_layout()
plt.savefig("figures/filter_size_comparison.png", dpi=300)
plt.show()


# ---------------------------------------------------
# 15. Experiment: compare pooling vs no pooling
# ---------------------------------------------------

pooling_results = []

for use_pooling in [True, False]:
    pooling_label = "With pooling" if use_pooling else "Without pooling"
    print(f"\nRunning pooling experiment: {pooling_label}")

    model = SimpleCNN(filter_size=3, use_pooling=use_pooling)

    trained_model, history = train_model(
        model,
        train_loader,
        val_loader,
        config,
        DEVICE,
        model_name=f"CNN (3x3, {pooling_label})"
    )

    test_loss, test_acc = evaluate(trained_model, test_loader, criterion, DEVICE)

    pooling_results.append({
        "model_name": f"CNN (3x3, {pooling_label})",
        "filter_size": "3x3",
        "pooling": "Yes" if use_pooling else "No",
        "parameters": count_parameters(trained_model),
        "test_loss": test_loss,
        "test_accuracy": test_acc,
        "main_observation": ""
    })

    # Save training curves for this variant as well.
    save_prefix = "with_pooling" if use_pooling else "without_pooling"
    plot_training_curves(
        history,
        model_name=f"CNN (3x3, {pooling_label})",
        save_prefix=save_prefix
    )

pooling_results_df = pd.DataFrame(pooling_results)
print("\nPooling comparison:")
print(pooling_results_df)

pooling_results_df.to_csv("results/pooling_results.csv", index=False)

# Plot test accuracy comparison for pooling variants.
plt.figure(figsize=(7, 5))
plt.bar(pooling_results_df["pooling"], pooling_results_df["test_accuracy"])
plt.xlabel("Pooling")
plt.ylabel("Test accuracy")
plt.title("Effect of Pooling on Test Accuracy")
plt.tight_layout()
plt.savefig("figures/pooling_comparison.png", dpi=300)
plt.show()


# ---------------------------------------------------
# 16. Create one final summary table
# ---------------------------------------------------

summary_rows = []

# Baseline row
summary_rows.append({
    "model_name": "Baseline CNN",
    "filter_size": "3x3",
    "pooling": "Yes",
    "parameters": count_parameters(baseline_model),
    "test_accuracy": baseline_test_acc,
    "main_observation": "Reference model used for feature-map and filter analysis."
})

# Add filter-size experiment rows
for _, row in filter_results_df.iterrows():
    observation = (
        "Smaller filters encourage deeper local composition."
        if row["filter_size"] == "3x3"
        else "Larger filters capture broader patterns earlier."
    )
    summary_rows.append({
        "model_name": row["model_name"],
        "filter_size": row["filter_size"],
        "pooling": row["pooling"],
        "parameters": row["parameters"],
        "test_accuracy": row["test_accuracy"],
        "main_observation": observation
    })

# Add pooling experiment rows
for _, row in pooling_results_df.iterrows():
    observation = (
        "Pooling improves spatial robustness and usually helps generalisation."
        if row["pooling"] == "Yes"
        else "Removing pooling preserves detail but may increase overfitting."
    )
    summary_rows.append({
        "model_name": row["model_name"],
        "filter_size": row["filter_size"],
        "pooling": row["pooling"],
        "parameters": row["parameters"],
        "test_accuracy": row["test_accuracy"],
        "main_observation": observation
    })

summary_df = pd.DataFrame(summary_rows)
summary_df.to_csv("results/final_summary_table.csv", index=False)

print("\nFinal summary table:")
print(summary_df)


# ---------------------------------------------------
# 17. Optional: save a cleaner markdown-style table
# ---------------------------------------------------

with open("results/final_summary_table.md", "w", encoding="utf-8") as f:
    f.write(summary_df.to_markdown(index=False))

print("\nSaved figures to ./figures and summary files to ./results")
```
